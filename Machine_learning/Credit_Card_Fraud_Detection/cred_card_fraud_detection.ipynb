# CREDIT CARD FRAUD DETECTION

### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### Load Dataset

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
# Check the number of missing values
credit_card_data.isnull().sum()
```

```python
# Distribution of legit transactions and fraudelent transactions
credit_card_data['Class'].value_counts()
```

0 ==> Normal Transaction

1 ==> Fraud Transaction

Fraud class count is very less. We cannot feed this data to our model.

This dataset is highly unbalanced

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape, fraud.shape)
```

```python
# Statistical analysis of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

### Under sampling
Build a sample dataset containing similar distribution of normal and fraudulent transactions

```python
legit_sample = legit.sample(n=492)
```

Concatenate 2 dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

### Split the data into Features and Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X.head()
```

```python
Y.head()
```

### Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
print(
	X_train.shape,
	X_test.shape,
	Y_train.shape,
	Y_test.shape
)
```

### Logistic Regression

```python
model = LogisticRegression(solver='lbfgs', max_iter=300)
```

```python
# Training the logistic regression model with the training data
model.fit(X_train, Y_train)
```

### Model Evaluation

```python
# Accuracy on Training Data
X_trian_prediction = model.predict(X_train)
trainig_data_accuracy = accuracy_score(X_trian_prediction, Y_train)
trainig_data_accuracy
```

```python
# Accuracy on Test Data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
test_data_accuracy
```
