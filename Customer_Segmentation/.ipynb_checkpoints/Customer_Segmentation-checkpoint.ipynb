## K-Means

### Age v/s Spending score clustering

```python
x1 = df.loc[:, ["Age", "Spending Score (1-100)"]].values

from sklearn.cluster import KMeans

wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters = k, init = "k-means++")
    kmeans.fit(x1)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (12, 6))
plt.grid()
plt.plot(range(1, 11), wcss, lw = 2, c = 'r', marker= '8')
plt.xlabel("K value")
plt.ylabel("WCSS")
plt.show()
```

```python
clusters_age = 4
```

```python
kmeans_age = KMeans(n_clusters = clusters_age)
```

```python
age_label = kmeans_age.fit_predict(x1)
print(age_label)
```

```python
print(kmeans_age.cluster_centers_)
```

```python
plt.scatter(x1[:, 0], x1[:, 1], c=kmeans_age.labels_, cmap='rainbow')
plt.scatter(kmeans_age.cluster_centers_[:, 0], kmeans_age.cluster_centers_[:, 1], color='black')
plt.title("Cluster of customers")
plt.xlabel("Age")
plt.ylabel("Spending score (1-100)")
plt.show()
```

### Annual Income v/s Spending Score clustering

```python
x2 = df.loc[:, ["Annual Income (k$)", "Spending Score (1-100)"]].values

from sklearn.cluster import KMeans

wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters = k, init = "k-means++")
    kmeans.fit(x2)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (12, 6))
plt.grid()
plt.plot(range(1, 11), wcss, lw = 2, c = 'r', marker= '8')
plt.xlabel("K value")
plt.ylabel("WCSS")
plt.show()
```

```python
ann_inc_clusters = 5
```

```python
kmeans_ann_inc = KMeans(n_clusters = ann_inc_clusters)

label_ann_inc = kmeans_ann_inc.fit_predict(x2)
print(label_ann_inc)
```

```python
print(kmeans_ann_inc.cluster_centers_)
```

```python
plt.scatter(x2[:, 0], x2[:, 1], c=kmeans_ann_inc.labels_, cmap='rainbow')
plt.scatter(kmeans_ann_inc.cluster_centers_[:, 0], kmeans_ann_inc.cluster_centers_[:, 1], color='black')
plt.title("Cluster of customers")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending score (1-100)")
plt.show()
```

## Clustering Using all features

```python
final_dataset = df.iloc[:, 1:]

wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters = k, init = "k-means++")
    kmeans.fit(final_dataset)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (12, 6))
plt.grid()
plt.plot(range(1, 11), wcss, lw = 2, c = 'r', marker= '8')
plt.xlabel("K value")
plt.ylabel("WCSS")
plt.show()
```

```python
final_clusters = 5
```

```python
kmeans = KMeans(n_clusters = final_clusters)

labels = kmeans.fit_predict(final_dataset)
print(labels)
```

```python
print(kmeans.cluster_centers_)
```

```python
clusters = kmeans.fit_predict(final_dataset)
df["label"] = clusters
```
